```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import datetime
from datetime import time

sns.set()
%matplotlib inline
```

```python
df = pd.read_csv("..\data\Parking_Citations_After_July_1_2015.csv", low_memory=False)
df.head()
```

```python
len(df)
```

```python
#remove the columns that I don't think are neccessary for this analysis
df= df.drop(["Ticket number",	"Meter Id",	"Marked Time",	"RP State Plate",	"Make",	"Body Style",	"Color",	"VIN",	"Violation code",	"Fine amount"], axis = 1)
#df= df.drop(["Route",	"Agency",	"Violation code",	"Violation Description",	"Fine amount"], axis = 1)
```

```python
df.head()
```

```python
x=(1850000<df['Longitude'])&(df['Longitude']<1875000)& (6490000<df['Latitude'])&(df['Latitude']<6520000)
```

```python
sdf=df[x]
sdf.head()
```

```python
len(sdf)
```

```python
 #create data Series num_sdf
num_sdf=sdf['Location'].value_counts()
```

```python
# reduce data series to be less unique values 
final_sdf = num_sdf[num_sdf>=50] # can be changed to 10 later on 
#list of index
final_sdf_index=final_sdf.index.tolist()
```

```python
final_index = sdf['Location'].isin(final_sdf_index)
```

```python
final_sdf = sdf[final_index]
```

```python
final_sdf.shape
```

```python
final_sdf.head()
```

```python
# convert location string to location ID 
final_sdf['Location']=final_sdf['Location'].astype('category')
final_sdf['Location_id']=final_sdf['Location'].cat.codes
final_sdf
```

```python
final_sdf['Issue Date'] = pd.to_datetime(final_sdf['Issue Date']) #change string to datetime
final_sdf['Time'] = pd.to_datetime(final_sdf['Issue time'], format= '%H%M', errors= "coerce").dt.time #change time in new column

#create an hour feature
final_sdf['Hour'] = final_sdf['Time'].apply(lambda x: x.hour) 
#fix converting bug
final_sdf['Hour'].fillna(0, inplace = True)

#create day of the week
weekdays=('Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday')
final_sdf['Weekday'] = final_sdf['Issue Date'].apply(lambda x : weekdays[x.weekday()])
final_sdf.head()
```

```python
p_sdf= final_sdf.drop(["Issue Date",	"Issue time",	"Location",	"Route",	"Agency",	"Violation Description",	"Latitude",	"Longitude",	"Time"], axis = 1)
```

```python
clean_sdf=p_sdf.dropna()
clean_sdf.apply(lambda x: sum(x.isna()),axis = 0)
clean_sdf.info()
```

```python
clean_sdf.head()
```

```python
filled_sdf = pd.get_dummies(clean_sdf, columns=['Weekday'])
filled_sdf.head()
```

```python
# Labels are the values we want to predict
labels = np.array(filled_sdf['Location_id'])
# Remove the labels from the features
# axis 1 refers to the columns
filled_sdf= filled_sdf.drop('Location_id', axis = 1)
# Saving feature names for later use
feature_list = list(filled_sdf.columns)
# Convert to numpy array
filled_sdf = np.array(filled_sdf)
```

```python
# Using Skicit-learn to split data into training and testing sets
from sklearn.model_selection import train_test_split
```

```python
# Split the data into training and testing sets
train_features, test_features, train_labels, test_labels = train_test_split(filled_sdf, labels, test_size = 0.3, random_state = 42)
```

```python
print('Training Features Shape:', train_features.shape)
print('Training Labels Shape:', train_labels.shape)
print('Testing Features Shape:', test_features.shape)
print('Testing Labels Shape:', test_labels.shape)
```

```python
# Import the model we are using
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, precision_score
```

```python
# Instantiate model with 1000 decision trees
#rf = RandomForestClassifier(n_estimators = 1000, random_state = 42)
# Train the model on training data
#rf.fit(train_features, train_labels)
```

```python
# Instantiate model with 1000 decision trees
rf = RandomForestClassifier(n_estimators = 1000, random_state = 42,max_depth=10,min_samples_leaf=10 )
# Train the model on training data
rf.fit(train_features, train_labels)
```

```python
prediction =rf.predict(test_features)
prediction
```

```python
from sklearn.metrics import make_scorer
from sklearn.metrics import log_loss
from sklearn.model_selection import GridSearchCV
model = RandomForestClassifier(n_estimators = 1000, random_state = 42)
depth_parm = [5, 10, 15, 20]
num_samples_parm = [5,8,10]
parameters = {'max_depth' : depth_parm,
             'min_samples_leaf' : num_samples_parm}
scorer = make_scorer(log_loss,
                     greater_is_better=False,
                     needs_proba=True)
clf = GridSearchCV(model,
                   parameters,
                   cv=10,
                   scoring=scorer)

clf.fit(train_features, train_labels)
#print(clf.best_params_)
```
