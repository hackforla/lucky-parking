```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import datetime
from datetime import time

sns.set()
%matplotlib inline

import geopandas
from geopandas import GeoDataFrame
from shapely.geometry import Point

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("..\data\Parking_Citations_After_July_1_2015.csv", low_memory=False)
df.head()
```

```python
geometry = [Point(xy) for xy in zip(df["Latitude"],df["Longitude"])] 
geometry[:3]
```

```python
#crs = {'init': 'epsg:4269'} #epsg:4362(WGS84) # crs= coordinate reference system 
df = geopandas.GeoDataFrame(df, geometry= geometry)
df.head()
```

```python
df=df[df['Latitude']!=99999]
df=df[df['Longitude']!=99999]
print(df['Latitude'].min(), df['Latitude'].max())
print(df['Longitude'].min(), df['Longitude'].max())
```

```python
#'lat_1': 34.03333333333333
```

```python
zipcode = geopandas.read_file('..\data\CAMS_ZIPCODE_STREET_SPECIFIC\CAMS_ZIPCODE_STREET_SPECIFIC.shp')
```

```python
zipcode.head()
```

```python
zip_df = geopandas.sjoin(df, zipcode, how="inner", op='intersects')
```

```python
zip_df.head()
```

```python
zip_df[['Location', 'Zip_Num']]
```

```python
zip_df.Zip_Num.value_counts()
```

```python
x=(1850000<zip_df['Longitude'])&(zip_df['Longitude']<1875000)& (6490000<zip_df['Latitude'])&(zip_df['Latitude']<6520000)
sdf=zip_df[x]
sdf.head()
```

```python
sdf.Zip_Num.value_counts()
```

```python
final_sdf=sdf[['Issue Date','Issue time','Zip_Num','Violation Description']]
final_sdf.head()
```

```python
final_sdf['Issue Date'] = pd.to_datetime(final_sdf['Issue Date']) #change string to datetime
final_sdf['Time'] = pd.to_datetime(final_sdf['Issue time'], format= '%H%M', errors= "coerce").dt.time #change time in new column

#create an hour feature
final_sdf['Hour'] = final_sdf['Time'].apply(lambda x: x.hour) 
#fix converting bug
final_sdf['Hour'].fillna(0, inplace = True)

#create day of the week
weekdays=('Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday')
final_sdf['Weekday'] = final_sdf['Issue Date'].apply(lambda x : weekdays[x.weekday()])
final_sdf.head()
```

```python
final_sdf['Violation Description'] = final_sdf['Violation Description'].astype('category')
final_sdf['Violation Description'] = final_sdf['Violation Description'].cat.codes
```

```python
final_sdf['Weekday'] = final_sdf['Weekday'].astype('category')
final_sdf['Weekday'] = final_sdf['Weekday'].cat.codes
```

final_sdf.drop(columns=['Issue Date', 'Issue time'], inplace = True)

```python
final_sdf.drop(columns=['Time'], inplace=True)
```

```python
final_sdf.drop(columns=['Violation Description'], inplace=True)
```

```python
final_sdf
```

```python
# Using Skicit-learn to split data into training and testing sets
from sklearn.model_selection import train_test_split
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(final_sdf[["Hour","Weekday"]], final_sdf['Zip_Num'], test_size = 0.2, random_state = 42)
```

```python
print('Training Features Shape:', X_train.shape)
print('Training Labels Shape:', X_test.shape)
print('Testing Features Shape:', y_train.shape)
print('Testing Labels Shape:', y_test.shape)
```

```python
# Import the model we are using
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, precision_score
```

```python
# Instantiate model with 1000 decision trees
rf = RandomForestClassifier(n_estimators = 1000, random_state = 42,max_depth=15,min_samples_leaf=20 )
# Train the model on training data
rf.fit(X_train, y_train)
```

```python
prediction =rf.predict(X_test)
prediction
```

```python
correct_prediction=prediction[y_test == prediction]
len(correct_prediction)
```

```python
len(prediction) - len(correct_prediction)
```

```python
len(correct_prediction)/len(prediction)
```

```python
from sklearn.metrics import make_scorer
from sklearn.metrics import log_loss
from sklearn.model_selection import GridSearchCV
model = RandomForestClassifier(n_estimators = 1000, random_state = 42)
depth_parm = [15, 20, 30, 40]
num_samples_parm = [20, 30 ]
inf = ['gini', 'entropy']
parameters = {'criterion' : inf,
    'max_depth' : depth_parm,
             'min_samples_leaf' : num_samples_parm}
#scorer = make_scorer(log_loss,
                    # greater_is_better=False,
                     #needs_proba=True)
clf = GridSearchCV(model,
                   parameters,
                   cv=5)

clf.fit(X_train, y_train)
print(clf.best_params_)
```
