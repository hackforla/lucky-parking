```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import datetime
from datetime import time

sns.set()
%matplotlib inline
```

```python
df = pd.read_csv("..\data\Simplified Citation After 2015 July to 2019.csv", low_memory=False)
df.head()
```

```python
df['Location']=df['Location'].astype('category')
df['Location_id']=df['Location'].cat.codes
df
```

```python
df['Location'].value_counts()
len(df['Location'].value_counts())
```

```python
df.info()
```

```python
# it takes lots time to process below amount of code, recommend to use simplified one 
df['Fine amount'] = pd.to_numeric(df['Fine amount'], errors="coerce") #change string to float
df['Issue Date'] = pd.to_datetime(df['Issue Date']) #change string to datetime
df['Time'] = pd.to_datetime(df['Issue time'], format= '%H%M', errors= "coerce").dt.time #change time in new column
```

```python
#split data into metered and non metered  
temp = {True:'Meter', False : 'No Meter'}
df['Meter']= (df['Meter Id'].notna()).map(temp)
 
#create an hour feature
df['Hour'] = df['Time'].apply(lambda x: x.hour) 

#create day of the week
weekdays=('Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday')
df['Weekday'] = df['Issue Date'].apply(lambda x : weekdays[x.weekday()])
```

```python
df['Year'] = df['Issue Date'].apply(lambda x : x.year)
df['Month'] = df['Issue Date'].apply(lambda x : x.month)
```

```python
#remove columns with too many none entries. this can only run once. otherwise startover again.
del df['VIN']
del df['Marked Time']
#remove the columns that I don't think are neccessary for this analysis
df= df.drop(["Ticket number",	"Meter Id",	"RP State Plate",	"Make",	"Body Style",	"Color"], axis = 1)
df= df.drop(["Route",	"Agency",	"Violation code",	"Violation Description",	"Fine amount",	"Latitude",	"Longitude",	"Issue Date",	"Time",	"Meter",	"Location"], axis = 1)
```

```python
df['Hour'].fillna(0, inplace = True)
df.head()
```

```python
df.info()
df.apply(lambda x: sum(x.isna()),axis = 0)
```

```python
df=df.dropna()
df.apply(lambda x: sum(x.isna()),axis = 0)
```

```python
#fill the data when a given location did not get citation related to time.
df = pd.get_dummies(df, columns=['Hour','Weekday'])
df.head()
```

```python
df.info()
```

```python
# Labels are the values we want to predict
labels = np.array(df['Location_id'])
# Remove the labels from the features
# axis 1 refers to the columns
df= df.drop('Location_id', axis = 1)
# Saving feature names for later use
feature_list = list(df.columns)
# Convert to numpy array
df = np.array(df)
```

```python
len(np.unique(labels))
```

```python
df
```

```python
# Using Skicit-learn to split data into training and testing sets
from sklearn.model_selection import train_test_split
```

```python
# Split the data into training and testing sets
train_features, test_features, train_labels, test_labels = train_test_split(df, labels, test_size = 0.8, random_state = 42)
```

```python
print('Training Features Shape:', train_features.shape)
print('Training Labels Shape:', train_labels.shape)
print('Testing Features Shape:', test_features.shape)
print('Testing Labels Shape:', test_labels.shape)
```

```python
# Import the model we are using
from sklearn.ensemble import RandomForestClassifier
```

```python
# Instantiate model with 1000 decision trees
rf = RandomForestClassifier(n_estimators = 1000, random_state = 42)
# Train the model on training data
rf.fit(train_features, train_labels)
```

```python
prediction =rf.predict(test_features)
prediction
```
