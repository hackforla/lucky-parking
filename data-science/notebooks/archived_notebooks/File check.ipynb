```python
df = pd.read_csv('Coordinates Converted_Simplified Citation After 2015 July to 2019.csv')
citation_locations = [citation['Location'] for index, citation in df.iterrows()]
info_box_template = """
<dl>
<dt>Violation</dt><dd>{Violation_Description}</dd>
<dt>Issue time</dt><dd>{Issue_time}</dd>
<dt>Issue Date</dt><dd>{Issue_Date}</dd>
</dl>
"""
citation_info = [info_box_template.format(**citation) for index, citation in df.iterrows()]

marker_layer = gmaps.marker_layer(citation_locations, info_box_content=citation_info)
fig = gmaps.figure()
fig.add_layer(marker_layer)
fig
```

```python
df['Location'] = (list(zip(df.Latitude, df.Longitude)))
```

```python
df.columns =  ['Issue_Date', 'Issue_time', 'Violation_Description', 'Day_of_Week',
       'Latitude', 'Longitude', 'Location']
```
